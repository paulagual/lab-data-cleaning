### Choose the 5 months: I'll keep the 5 months with more data

Choose the rows with less than 29 Nulls in both tmin and tmax

```python
data_5 = data[data.T.isnull().sum()<29]
```

```python
data_5
```

### Melt the table to have the days in the rows

```python
melted = pd.melt(data_5, id_vars=['id','year','month','element'], 
                 value_vars=['d1', 'd2', 'd3', 'd4', 'd5', 'd6',
       'd7', 'd8', 'd9', 'd10', 'd11', 'd12', 'd13', 'd14', 'd15', 'd16',
       'd17', 'd18', 'd19', 'd20', 'd21', 'd22', 'd23', 'd24', 'd25', 'd26',
       'd27', 'd28', 'd29', 'd30', 'd31'])
```

```python
melted.head()
```

### Check the null values of the melted dataframe and drop the null rows

```python
empty_rows = list(melted[melted['value'].isnull() == True].index)
```

```python
melted = melted.drop(melted.index[empty_rows])
```

```python
melted.head()
```

### Pivot the melted table to "unmelt" the dataframe

```python
unmelted = melted.pivot_table(index=['id','year', 'month'], 
                              columns='element')
```

```python
unmelted.columns = unmelted.columns.droplevel().rename(None)
```

```python
unmelted.columns
```

```python
unmelted = unmelted[['tmin', 'tmax']]
```

```python
unmelted
```

Round the results of the unmelted dataframe

```python
unmelted = unmelted.round(2)
```

```python
unmelted
```

### Temp Means by month and element without the NaN Values

```python
avg_temp = data_5[['d1', 'd2', 'd3', 'd4', 'd5', 'd6',
       'd7', 'd8', 'd9', 'd10', 'd11', 'd12', 'd13', 'd14', 'd15', 'd16',
       'd17', 'd18', 'd19', 'd20', 'd21', 'd22', 'd23', 'd24', 'd25', 'd26',
       'd27', 'd28', 'd29', 'd30', 'd31']].mean(axis = 1, skipna = True) 
```

```python
avg_temp.name = 'Mean Temp'
```

```python
avg_temp
```

### Add the mean row to the main dataframe

```python
data_5 = pd.concat([data_5, avg_temp], axis=1)
```

```python
data_5.head()
```

### Create a new data frame with month, element and mean

```python
short_data = data_5[['id', 'year', 'month', 'element', 'Mean Temp']]
```

```python
short_data
```

### Create a DataFrame of months and Mean of Min Temp

```python
short_data_min = short_data.loc[short_data['element'] == 'tmin']
```

```python
short_data_min
```

```python
short_data_min.columns
```

```python
short_data_min.columns = ['id', 'year', 'month', 'element', 'mean tmin']
```

```python
short_data_min = short_data_min.drop(['element'], axis=1)
```

```python
short_data_min
```

### Create a DataFrame of months and Mean of Max Temp


```python
short_data_max = short_data.loc[short_data['element'] == 'tmax']
```

```python
short_data_max
```

```python
short_data_max.columns = ['id', 'year', 'month', 'element', 'mean tmax']
```

```python
short_data_max = short_data_max.drop(['element'], axis=1)
```

```python
short_data_max
```

### Create a results DataFrame 

```python
result = pd.merge(short_data_min, short_data_max, on=['id', 'year', 'month'])
```

### Round the result dataframe

```python
result = result.round(2)
```
